# How to analyse Google trends using Pytrends library

Installing Pytrends library

Create a TrendReq object

Defining the hl parameter as the host language for accessing Google Trends

speechemotion environment

```python
from pytrends.request import TrendReq
import seaborn

seaborn.set_style("whitegrid")

pt = TrendReq(hl="fi-FI", tz=360)
```

Let's use the interest_over_time() method to analyse the search trends concerning Elisa Viihde and Telia TV.

The values range from 0 which means few or no searches to 100 maximum possible searches.

```python
pt.build_payload(["Elisa Viihde", " Telia TV"], timeframe="all")

iot = pt.interest_over_time()
iot
```

Let's plot the relative search difference between Elisa Viihde and Telia TV over time:

```python
import matplotlib.pyplot as plt

iot.plot(figsize=(10, 6))
plt.savefig("elisatelia.png")
```

```python
pt.build_payload(["Tyllisyysrahasto", "Aikuiskoulutustuki"], timeframe="all")

iot1 = pt.interest_over_time()
iot1
```

```python
import matplotlib.pyplot as plt

iot1.plot(figsize=(10, 6))
plt.savefig("tyollisyysrahasto.png")
```

Let's get the interest of a specific keyword such as Tesla by region:

```python
kw = "Tesla"
pt.build_payload([kw], timeframe="all")
ibr = pt.interest_by_region("COUNTRY", inc_low_vol=True, inc_geo_code=True)
```

Let's sort the countries by interest in Tesla:

```python
ibr[kw].sort_values(ascending=False)
ibr
```

Let's plot the top 10 countries by interest.

```python
import matplotlib.pyplot as plt
ibr[kw].sort_values(ascending=False)[:10].plot.bar()
plt.savefig("teslasanamaakehitys.png")
```

The related_topics() method returns a Python dictionary of each keyword such as in this case Tesla; this dictionary has two dataframes, one for rising topics and one for overall top topics.

```python
rt = pt.related_topics()
rt[kw]["top"]
```

Let's see first the top topics.

```python
rq = pt.related_queries()
rq[kw]["top"]
```

Let's see then the rising topics.

```python
rq = pt.related_queries()
rq[kw]["rising"]
```

One more feature on Google trends is the ability to extract the current trending searches on each region:

```python
ts = pt.trending_searches(pn="finland")
ts[:50]
```

Let's see what is the interest search rate for Espoo and Vantaa?

```python
from pytrends.request import TrendReq
import plotly.express as px 

pytrends = TrendReq(hl='fi-FI', tz=360) 
```

```python
kw_list = ["Espoo", 'Vantaa'] 

pytrends.build_payload(kw_list, cat=0, timeframe='today 12-m') 
```

```python
import matplotlib.pyplot as plt
data = pytrends.interest_over_time() 
data = data.reset_index()


fig = px.line(data, x="date", y=kw_list, title='Hakusanojen kehitys 2021-2022')
fig.show() 
plt.savefig("avainsanojenkehitys.png")
```
